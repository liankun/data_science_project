### Best zip code 
This place has more markets  and a higher population. It looks like downtown. For me, it is a good place to open a bubble tea shop.

```python
cand_zipcode = 38017
labels=[]
deviation=[]
for col in predict1_df.columns:
    if col == 'Unnamed: 0' or col == 'zip' or col == "LABEL_bubbleStore_cnt_fst_agg":
        continue
    labels.append(col)
    val = predict1_df[predict1_df['zip']==cand_zipcode][col].values[0]
    m = predict1_df[col].mean()
    d = predict1_df[col].std()
    if d!=0:
        deviation.append((val-m)/d)
    else:
        deviation.append(0)
    #print(col)
    #print(val)
plt.figure(figsize=(10,10))
plt.barh(labels,deviation)
```
