## Statistical test

```python
from scipy.stats import mannwhitneyu
```

```python
U_ls = []
p_ls = []
cols = []
```

```python
for i in range(intersec.shape[1]):
    U1, p = mannwhitneyu(intersec.iloc[:,i], intersec_poor.iloc[:,i], method="exact")
    print(intersec.columns[i])
    cols.append(intersec.columns[i])
    U_ls.append(U1)
    p_ls.append(p)
    print(U1, p)
```

```python
from matplotlib import pyplot as plt
from matplotlib import *
import numpy as np
sort_idx = list(np.argsort(U_ls[1:]))

# Set the figure size
plt.rcParams["figure.figsize"] = [10.00, 7]
# plt.rcParams["figure.autolayout"] = True

x = [cols[1:][i] for i in sort_idx]
y = [U_ls[1:][i] for i in sort_idx]
# Plot bar chart with data points
plt.bar(x, y)
plt.xticks(x, rotation=90)
plt.ylabel('Wilcoxon-Mann-Whitney test U Score')
# Display the plot
plt.show()
```
